### Question 1

```python
print("First 10 prime numbers:")

count = 0
num = 2

while count < 10:
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")
        count += 1
    num += 1
```

### Question 5

```python
features_keep = [
    "Daily new confirmed cases of COVID-19",
    "Daily new confirmed deaths due to COVID-19",
    "Total confirmed cases of COVID-19",
    "Total confirmed deaths due to COVID-19",
    "Case fatality rate of COVID-19 (%)",
    "Weekly cases",
    "Weekly deaths",
    "Weekly case growth (%)",
    "Weekly death growth (%)"
]
covid19_sa = covid19_sa[features_keep]
covid19_sa.head()
```

### Question 6

```python
numeric_cols = covid19_sa.select_dtypes(include=['number']).columns
covid19_sa[numeric_cols] = covid19_sa[numeric_cols].fillna(covid19_sa[numeric_cols].mean())
covid19_sa = covid19_sa.drop_duplicates()
covid19_sa.info()
```

### Question 7

```python
Avg_Deaths = covid19_sa["Daily new confirmed deaths due to COVID-19"].mean()
print("Average daily new confirmed deaths in South Africa:",round(Avg_Deaths,2))
```

### Question 8

```python
Avg_Daily_Confirmed_Cases = covid19_sa["Daily new confirmed cases of COVID-19"].mean()
print("Average daily new confirmed cases of COVID-19 in South Africa:", round(Avg_Daily_Confirmed_Cases, 2))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(Avg_Daily_Confirmed_Cases, Avg_Deaths, color='Green')
plt.title("Average Daily COVID-19 Cases against Average Daily Deaths (South Africa)")
plt.xlabel("Average Daily New Confirmed Cases")
plt.ylabel("Average Daily New Confirmed Deaths")
plt.grid(True)
plt.show()
```
